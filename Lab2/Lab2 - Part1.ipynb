```
from random import randrange
s1= [randrange(0, 100) for _ in range(0, 100)]
s2= [randrange(0, 100) for _ in range(0, 100)]

df1 = pd.DataFrame(s1)
df2 = pd.DataFrame(s1)
```

```
c1 = df1 + df2
c2 = df1 - df2
c3 = df1 * df2
c4 = df1 / df2
```

```
db = []
for i in range(0, 100):
  db.append({
      'Num1':s1[i],
      'Num2':s2[i],
      'Addition':c1[0][i],
      'Subtraction':c2[0][i],
      'Multiplication':c3[0][i],
      'Division':c4[0][i],
  })
db = pd.DataFrame(db)
db.head()
```
