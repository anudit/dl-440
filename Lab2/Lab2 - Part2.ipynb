```
import pandas as pd
df = pd.read_csv('Advertising.csv')
df.head()
```

```
import numpy as np
df['Sales Type'] = np.where(df['Sales'] < 15, np.where(df['Sales'] < 5, 'Low', 'Medium'), 'High')
df.head()
```

```
print(df.groupby(['Sales Type']).sum())
print(df.groupby(['Sales Type']).mean())
```

```
df.pivot(index='TV', columns='Radio', values='Radio')
```

```
for index, row in df.iterrows():
  if (index != 'TV' and int(index)%6 == 0):
    df.iloc[index]['TV'] = 'NULL'
for index, row in df.iterrows():
  if (index != 'TV' and int(index)%10 == 0):
    df.iloc[index]['Newspaper'] = float(0)
```
